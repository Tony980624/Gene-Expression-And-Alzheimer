```python
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(r'D:\STATA\Unormalized_df.csv',index_col=0)
df
```

```python
X = df.drop(columns=['Disease'])
y = df['Disease']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=47616554)
```

```python
model = GradientBoostingClassifier(n_estimators=1000, max_depth=4, random_state=47616554)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')

print("Classification Report:")
print(classification_report(y_test, y_pred))


print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import numpy as np


feature_importances = model.feature_importances_

feature_names = X.columns
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importances})

feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

print(feature_importance_df.head(10))  # 查看前 10 个重要特征

plt.figure(figsize=(10, 6))
plt.barh(feature_importance_df['Feature'][:10], feature_importance_df['Importance'][:10])
plt.gca().invert_yaxis()
plt.xlabel("Importance")
plt.title("Top 10 Feature Importances")
plt.show()
```

```python
top_features = ["10025903459", "10023824754", "10025910422"]
df.groupby("Disease")[top_features].describe()
```

```python
for feature in top_features:
    group0 = df[df['Disease'] == 0][feature]
    group1 = df[df['Disease'] == 1][feature]

    mean0, mean1 = np.mean(group0), np.mean(group1)
    var0, var1 = np.var(group0, ddof=1), np.var(group1, ddof=1)  
    n0, n1 = len(group0), len(group1)
    
    t_stat = (mean0 - mean1) / np.sqrt(var0 / n0 + var1 / n1)
    
    print(f"T-statistic for {feature}: {t_stat:.4f}")
```

```python
# Apparent error rate
model.fit(X, y)
apparent_predictions = model.predict(X)


apparent_accuracy = accuracy_score(y, apparent_predictions)
apparent_error_rate = 1 - apparent_accuracy
print(f"Apparent Error Rate: {apparent_error_rate:.4f}")
```

```python
# Cross validation

from sklearn.model_selection import cross_val_predict
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


cv_predictions = cross_val_predict(model, X, y, cv=5)

overall_accuracy = accuracy_score(y, cv_predictions)
overall_error_rate = 1 - overall_accuracy
print(f"Overall CV-based Error Rate: {overall_error_rate:.4f}")


print("Class-Specific Error Rates:")
print(classification_report(y, cv_predictions))

```
